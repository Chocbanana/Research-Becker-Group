## Turn video into series of images

````python
import cv2
from matplotlib import pyplot as plt
import numpy as np
````

````python
# img_size = (640, 360)
img_size = (96, 54)

def vid_to_imgs(vid_name: str, img_folder: str, img_name: str):
    video = cv2.VideoCapture(f"../data/{vid_name}")

    i = 0
    hasframe, frame = video.read()

    while (hasframe):
        cv2.imwrite(f"../data/{img_folder}/{img_name}_{i}.jpg", cv2.resize(frame, img_size, interpolation= cv2.INTER_LINEAR))
        hasframe, frame = video.read()
        i += 1
        
    print(f"Made {i} images")
````

````python
vid_to_imgs("test_video1.webm", "test_imgs0", "imageA")
vid_to_imgs("test_video2.mp4", "test_imgs1", "imageB")
vid_to_imgs("test_video3.webm", "test_imgs2", "imageC")
````

# Make Movie From Matlab Data

````python
import os

from matplotlib import pyplot as plt
from matplotlib.animation import ArtistAnimation, FFMpegWriter
import numpy as np
````

````python
data_path = os.path.normpath("../data/")
plasma_n64_path = os.path.join(data_path, "gen_plasma_n128")
folder = "mat_1"
n_img = 16294
mats_path = os.path.join(plasma_n64_path, folder)

fig, ax = plt.subplots()
ax.set_axis_off()

imgs = []
for i in range(n_img):
    mat = np.genfromtxt(os.path.join(mats_path, f"f_{i}.csv"), delimiter=",", dtype=float)
    img = ax.imshow(mat, animated=True)
    imgs.append([img])
````

````python
video = ArtistAnimation(fig, imgs, repeat=False)
video.save(os.path.join(plasma_n64_path, f"{folder}_f.mp4"), fps=120)
````
